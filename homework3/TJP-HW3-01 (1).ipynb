```python
import requests
import bs4
import csv
from bs4 import BeautifulSoup
```

## Scrape the data of job

```python
def all(n):
        r=requests.get(n)
        r.encoding=('utf8')
        html_str = r.text
        mypage = BeautifulSoup(html_str,'html.parser')
        myheadline = mypage.find_all('span',attrs ={'class':'name'})
        job = []
        for m in myheadline:
            job.append(m.text.split()[0])
        return job
```

```python
page_next1 = []
for q in range (1,31):
    n = 'http://107cine.com/job/home?v_zhiwei=%E5%AE%9E%E4%B9%A0&city=all&page=' + str(q+1)
    page_next1.extend(all(n))
    
```

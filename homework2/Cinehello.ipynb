```python
import requests
import bs4
import csv
```

```python
r=requests.get('http://107cine.com/job/home/?v_zhiwei=%E5%AE%9E%E4%B9%A0&city=45#ads-banner')
```

```python
r.text
```

```python
from bs4 import BeautifulSoup
```

```python
mypage=BeautifulSoup(r.text)
```

```python
mypage
```

```python
myheadline=mypage.find_all('span',attrs={'class':'name'})
```

```python
myheadline
```

```python
t=[]
for x in myheadline:
    h = x.find('a')
    t.append(h.text)
```

```python
t
```

```python
mysalary=mypage.find_all('span',attrs={'class':'pay'})
```

```python
mysalary
```

```python
list_mysalary = []
for salary in mysalary:
    salary.text
    list_mysalary.append(salary.text)
```

```python
list_mysalary
```

```python
with open('CINEdata.csv','w') as f:
    writer = csv.writer(f)
    header = ['t','list_mysalary']
    writer.writerow(header)
    
    writer.writerows(zip(t,list_mysalary))
```
